```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('messi.jpg')  #read image
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #convert image from BGR to RGB
plt.imshow(img) #display image
plt.show()
```

```python
#translation
T=np.float32([[1,0,100],[0,1,100],[0,0,1]]) #create translation matrix
trans_img=cv2.warpPerspective(img,T,(img.shape[1],img.shape[0])) #create new image with translation by 100,100
plt.imshow(trans_img)
plt.show()
```

```python
#Rotation
angle=np.deg2rad(30) #create a radian rotation angle 
R=np.float32([[np.cos(angle),-(np.sin(angle)),0],[np.sin(angle),np.cos(angle),0],[0,0,1]]) #create rotation matrix
rot_img=cv2.warpPerspective(img,R,(img.shape[1],img.shape[0])) #create new image with rotation by 30
plt.imshow(rot_img)
plt.show() 
```

```python
#scaling
S=np.float32([[1.5,0,0],[0,1.5,0],[0,0,1]]) #create scaling matrix
scale_img=cv2.warpPerspective(img,S,(img.shape[1],img.shape[0])) #create new image with scaling by 1.5,1.5
plt.imshow(scale_img)
plt.show()
```
